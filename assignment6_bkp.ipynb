```python
#Change kernel to python 

from collections import Counter
import matplotlib.pyplot as plt


def get_token_tallies(file):
    with open(file, 'r') as f:
        lines = f.readlines()
    for i in range(len(lines)):
        lines[i] = lines[i].strip()
    print("totallines: ", len(lines), "unique", len(set(lines)))
#     print(set(lines))
    return lines
    
def plot_hist(file_with_lines):
    plt.rcParams["figure.figsize"] = (20,3)
    w = Counter(file_with_lines)
    top = w.most_common()
    keys = [top[i][0] for i in range(len(top))]
    values = [top[i][1] for i in range(len(top))]
    plt.bar(keys, values)
    plt.show()
```

# Swara Phrases

```python
swara_vocab = [ 'S', 'R', 'G', 'M', 'P', 'D', 'N',
                's','r','g','m','p','d','n', 
                'ss','rr','gg','mm','pp','dd','nn', 'hr','hN' ',', ')', '('
              ]
```

```python
import re 
with open('Analysis/clean_files_final/by_rasas/adbhut/Alhaiya_bilaval_D1T4.txt', 'r') as f:
    file_lines = f.readlines()
for i in range(len(file_lines)):
        file_lines[i] = file_lines[i].strip()
        
phrases = []
for ind, word in enumerate(file_lines):

    patrn = r"[*,]"
    if re.search(patrn, word):
        continue
    else:
        patrn2 = r"[\(]"  #phrase start
        patrn3 = r"[.*\)]" #phrase end
        if re.search(patrn2, word):
            print('phrase_start')

            new_phrase = ['(']
            continue
        elif (re.search(patrn3, word) == None):
            if word in swara_vocab:
                print(ind,word, 'word in swara vocab')
                new_phrase.append(word)
                continue
            else:
                print(ind, word, 'word NOT in swara vocab')
                continue
        elif (word[:-1]) in swara_vocab:
                new_phrase.append(word[:-1])
                new_phrase.append(')')
                phrases.append(new_phrase)            
```
